# 2. 索引

### 将某列设置为索引

```python
# Setting a new index from an existing column
df = df.set_index(['water_year'])
df.head(5)
```

### 将索引排序

```python
df.sort_index(ascending=False).head(5) 
```

### 按某列的值排序

```python
df.sort_values(by = 'rain_octsep').head(5)
```

### 将索引恢复为数据

```python
# Returning an index to data
df = df.reset_index()
df.head(5)
```

# 3. 对数据集应用函数
有时你想对数据集中的数据进行改变或者某种操作。
- 比方说，你有一列年份的数据，你需要新的一列来表示这些年份对应的年代。
- Pandas 中有两个非常有用的函数，**apply** 和 **applymap**。

```python
# Applying a function to a column
def fun(year):
    base_year = year[:4]
    base_year= pd.to_datetime(base_year).year
    return base_year

df['year'] = df.water_year.apply(fun)
df.head(5)
```

这就是 apply 的用法，即对一列数据应用函数。如果你想对整个数据集应用函数，就要使用 applymap。

# 4. 操作数据集的结构
另一常见的做法是重新建立数据结构，使得数据集呈现出一种更方便并且（或者）有用的形式

### 首先是 groupby

```python
#Manipulating structure (groupby, unstack, pivot)
# Grouby
df.groupby(df.year // 10 *10).max()
```

```python
# Grouping by multiple columns
decade_rain = df.groupby([df.year // 10 * 10, df.rain_octsep // 1000 * 1000])[['outflow_octsep', 'outflow_decfeb', 'outflow_junaug']].mean()
decade_rain
```

---
### 接下来是 unstack

```python
# Unstacking
decade_rain.unstack(0)
```

```python
# More unstacking
decade_rain.unstack(1)
```

# 5. 轴向旋转（pivoting）
在进行下次操作之前，我们先创建一个用于演示的 dataframe :

```python
# Create a new dataframe containing entries which 
# has rain_octsep values of greater than 1250
high_rain = df[df.rain_octsep > 1250]
high_rain
```

--
轴旋转其实就是我们之前已经看到的那些操作的一个集合。首先，它会设置一个新的索引（set_index()），然后对索引排序（sort_index()），最后调用 unstack 。以上的步骤合在一起就是 pivot。
下面的 dataframe 展示了所有降雨超过 1250 的 outflow

```python
#Pivoting
#does set_index, sort_index and unstack in a row
high=high_rain.pivot('year', 'rain_octsep')[['outflow_octsep', 'outflow_decfeb', 'outflow_junaug']].fillna('')
high
```

---
# 6. 保存你的数据集

```python
high.to_csv('uk_rain.csv')
```
